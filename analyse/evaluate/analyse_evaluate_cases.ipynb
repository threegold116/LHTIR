对 `results/` 下 BFCL 等评测产物做逐条检查。`*_result.json`、`*_score.json` 在本仓库中一般为 **JSONL**（每行一个 JSON 对象），与 `analyse/rollout/analyse_rollout_cases.ipynb` 的用法类似。

逻辑封装在同级脚本 `analyse_evaluate_cases.py` 中，也可直接运行：

```bash
python analyse/evaluate/analyse_evaluate_cases.py --root /path/to/results --suffix _result.json
python analyse/evaluate/analyse_evaluate_cases.py --file path/to/BFCL_v4_xxx_result.json --print-structure
```
